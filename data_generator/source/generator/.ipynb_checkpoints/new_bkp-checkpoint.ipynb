```python
from treelib import Node, Tree
import csv
from concurrent.futures import ThreadPoolExecutor
import random
import os
import json
import pprint
```

```python
my_dict = {
    "level1": {
        "is_parent": True,
        "number_of_distinct_vals": 0,
        "total_vals": 0,
        "dups": 2
        #,
        #"key_2": "test2"
    },
    "level2": {
        "is_parent": True,
        "number_of_distinct_vals": 0,
        "total_vals": 0,
        "dups": 0
        #,
        #"key_2": "test2"
    },
    "level3": {
        "is_parent": False,
        "number_of_distinct_vals": 2,
        "total_vals": 0, #number of distinct values * number of dups of last level,
        "dups": 3
        #,
        #"key_2": "test2"
    },
    "level4": {
        "is_parent": False,
        "number_of_distinct_vals": 2,
        "total_vals": 0,
        "dups": 2
        #,
        #"key_2": "test2"
    },
    "level5": {
        "is_parent": False,
        "number_of_distinct_vals": 0,
        "total_vals": 0,
        "dups": 2
        #,
        #"key_2": "test2"
    },
    "level6": {
        "is_parent": False,
        "number_of_distinct_vals": 0,
        "total_vals": 0,
        "dups": 2
        #,
        #"key_2": "test2"
    },
    "level7": {
        "is_parent": True,
        "number_of_distinct_vals": 0,
        "total_vals": 0,
        "dups": 0
        #,
        #"key_2": "test2"
    },
    "level8": {
        "is_parent": False,
        "number_of_distinct_vals": 0,
        "total_vals": 0,
        "dups": 0
        #,
        #"key_2": "test2"
    }
}
```

```python
def update_input(current_level=1,first_level=8):
    print(current_level)
    v_add_up=1
    v_distinct_values=2
    if my_dict["level"+str(current_level)]["is_parent"]==False and my_dict["level"+str(current_level)]["total_vals"]==0 and current_level+1<=first_level:
        v_add_up=v_add_up*my_dict["level"+str(current_level)]["number_of_distinct_vals"]*update_input(current_level+1)
        if my_dict["level"+str(current_level+1)]["is_parent"]==True:
            v_add_up=v_add_up*my_dict["level"+str(current_level)]["dups"]
            my_dict["level"+str(current_level)]["total_vals"]=v_add_up
        else:
            my_dict["level"+str(current_level)]["total_vals"]=v_add_up
    elif my_dict["level"+str(current_level)]["is_parent"]==False and my_dict["level"+str(current_level)]["total_vals"]==0 and current_level+1>first_level:
        my_dict["level"+str(current_level)]["total_vals"]=v_distinct_values
    pp = pprint.PrettyPrinter(indent=4)
    pp.pprint(my_dict)
    return v_add_up 
```

```python
first_level=8
for i in range(1,first_level+1):
    update_input(i,first_level)
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(my_dict)
```

```python
def generate_data(level=0,first_level=7,tree=Tree(),parent="",parent_index=0,letter=65,num_of_dups=3,out_text="",dup_sink_rate=1,start=1,end=3,FD_dict={},addup=0):
    for i in range(1,first_level+1):
        update_input(i,first_level)
    if level!=0:
        num_of_dups=my_dict["level"+str(level)]["dups"]
        is_parent=my_dict["level"+str(level)]["is_parent"]  #it should be automatic (as input)
        if my_dict["level"+str(level)]["dups"]!=0:
            dup_enabled=True
        else:
            dup_enabled=False #it should be automatic (as input)
    number_of_dup_values=2 #it should be clacualted (according to some formula based on the total number of records)
    number_of_values=2 #it should be clacualted (according to some formula based on the total number of records)
    v_letter=65
    if level==first_level:
        if is_parent==False:
            for i in range(1,3):
                tree.create_node(chr(letter)+str(i), out_text+chr(letter)+str(i),parent)
                new_file="tree.csv"
                with open(new_file, 'a+',newline="") as f_out:
                    f_out.write(out_text+chr(letter)+str(i)+"\n")
        else:
            for i in range(start,end):
                tree.create_node(chr(letter)+str(i),out_text+chr(letter)+str(i),parent)
                new_file="tree.csv"
                with open(new_file, 'a+',newline="") as f_out:
                    f_out.write(out_text+chr(letter)+str(i)+"\n") 
    else:
        if level!=0:
            if level==1:
                for i in range(1,number_of_values+1):
                    start_point=(i-1)*num_of_dups+1
                    if i>number_of_dup_values:
                        end_point=(i-1)*num_of_dups+2
                    else:
                        end_point=(i-1)*num_of_dups+num_of_dups+1
                    out_text=out_text.replace(chr(letter)+str(i-1)+",","")
                    out_text=out_text+chr(letter)+str(i)+","
                    tree.create_node(chr(letter)+str(i),out_text,parent)
                    generate_data(level+1,first_level,tree,parent=out_text,parent_index=i,letter=v_letter+first_level-level-1,num_of_dups=int(num_of_dups*dup_sink_rate),out_text=out_text,start=start_point,end=end_point)
            else:
                if is_parent==False:
                    num_of_distinct_vals=2
                    for i in range(1,num_of_distinct_vals+1):
                        v_addup=addup+(parent_index-1)*my_dict["level"+str(level)]["total_vals"]
                        if dup_enabled==False:
                            start_point=((i-1)*num_of_dups)+1
                            end_point=((i-1)*num_of_dups)+2
                        else:
                            start_point=v_addup+(i-1)*num_of_dups+1
                            end_point=v_addup+i*num_of_dups+1
                            if level==6:
                                print(start_point)
                        out_text=out_text.replace(chr(letter)+str(i-1)+",","")
                        out_text=out_text+chr(letter)+str(i)+","
                        tree.create_node(chr(letter)+str(i),out_text,parent)
                        generate_data(level+1,first_level,tree,parent=out_text,parent_index=i,letter=v_letter+first_level-level-1,num_of_dups=int(num_of_dups*dup_sink_rate),out_text=out_text,start=start_point,end=end_point,addup=v_addup)
                else:
                    for i in range(start,end):
                        if dup_enabled==False:
                            start_point=((i-1)*num_of_dups)+1
                            end_point=((i-1)*num_of_dups)+2
                        else:
                            start_point=((i-1)*num_of_dups)+1
                            end_point=((i-1)*num_of_dups)+num_of_dups+1
                        out_text=out_text.replace(chr(letter)+str(i-1)+",","")
                        out_text=out_text+chr(letter)+str(i)+","
                        tree.create_node(chr(letter)+str(i),out_text,parent)
                        generate_data(level+1,first_level,tree,parent=out_text,parent_index=i,letter=v_letter+first_level-level-1,num_of_dups=int(num_of_dups*dup_sink_rate),out_text=out_text,start=start_point,end=end_point,addup=0)
        else:
            #os.remove("tree.csv")
            tree.create_node("root", "root")
            generate_data(level+1,first_level,tree,parent="root",letter=v_letter+first_level-level-1)
    if level==0:
        print(tree.show())
```
