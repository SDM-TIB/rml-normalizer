```python
class Tree(object):
    def __init__(self):
        self.left = None
        self.child = []
        self.data = []
    def createChildren(self,amount):
        for i in range(0,amount):
            self.child.append(Tree())
    def setChildrenValues(self,list):
        for i in range(0,len(list)):
            self.data.append(list[i])
```
